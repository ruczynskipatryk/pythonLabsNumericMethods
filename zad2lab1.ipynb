```python
import numpy as np

xp = 0
xk = 2*np.pi
krok = 0.1
liczba_punktow = np.abs((xk-xp)/krok) # wyznacz 1. punktów
liczba_punktow_int = int(np.ceil(liczba_punktow)) # zadbaj o zgodność typów
print("liczba punktów wartości x: %d" % liczba_punktow)
x = xp
for i in range(0,liczba_punktow_int):
    x = x+krok
    print("sin(%f)=%f" % (x,np.sin(x)))
```
