```python
"""
    import biblioteki NumPy do przestrzeni nazw np;
    funkcje dostepne są z poziomu przestrzeni nazw 'np'
"""

import numpy as np
x=np.pi/3 #twórz zmienną x
#oblicz wartości f trygonometrycznyc
print("sin(%f)=%f" % (x,np.sin(x)))
print("sin(%f)=%f" % (x,np.cos(x)))
print("tan(%f)=%f" % (x,np.tan(x)))
print("ctg(%f)=%f" % (x,1.0/np.sin(x)))
```
